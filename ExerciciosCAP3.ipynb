```python
#3.9
dias = int(input('Digite a quantidade de dias: '))
horas = int(input('Digite a quantidade de horas:'))
minutos = int(input('Digite a quantidade de minutos: '))
segundos = int(input('Digite a quantidade de segundos '))

segundos_dos_dias = dias * 86400
segundos_das_horas = horas * 3600
segundos_dos_minutos = minutos * 60
total_em_segundos = segundos_dos_dias + segundos_das_horas + segundos_dos_minutos + segundos

print(total_em_segundos)
```

```python
#Ex3.15
quantidade_cigarros = int(input('Quantos cigarros fuma por dia?: '))
anos_fumando = float(input('Digite a quantidade de tempo que fuma em anos: '))
dias_fumando = anos_fumando * 365
total_cigarros = quantidade_cigarros * dias_fumando
minutos_perdidos = total_cigarros * 10
tempo_perdido = minutos_perdidos / (24 * 60)
print(f'{tempo_perdido:.2f}')
```
