```python
#Programa 4.3
salario = float(input('Digite o salário para cálculo do imposto: '))
base = salario
imposto = 0
if base > 3000:
  imposto = imposto + ((base - 3000) * 0.35)
  base = 3000
if base > 1000:
  imposto = imposto + ((base - 1000) * 0.20)

print(f'Salário: R${salario:6.2f}. Imposto a pagar: R${imposto:6.2f}')
```

```python
#4.8
escolha = input('Digite o tipo de operação que deseja realizar: ')


numero1 = float(input('Digite o primeiro número: '))
numero2 = float(input('Digite o segundo número: '))


if escolha == 'soma':
  valor_somado =  numero1 + numero2
  print(f'A soma do {numero1} + {numero2} é: {valor_somado}')

elif escolha == 'subtraçao':
  valor_somado =  numero1 - numero2
  print(f'A soma do {numero1} - {numero2} é: {valor_somado}')

elif escolha == 'multiplicaçao':
  valor_somado =  numero1 * numero2
  print(f'A soma do {numero1} * {numero2} é: {valor_somado}')

elif escolha == 'divisao':
  valor_somado =  numero1 / numero2
  print(f'A soma do {numero1} / {numero2} é: {valor_somado}')

else:
  print('Operação não existe! Tente novamente!!!')
```

```python
#4.10
kwh = float(input("Digite a quantidade de kWh consumida: "))
tipo = input("Digite o tipo de instalação (R para residência, I para indústria, C para comércio): ").upper()

preco_kwh = 0
valido = True

if tipo == 'R':
    if kwh <= 500:
        preco_kwh = 0.40
    else:
        preco_kwh = 0.65

elif tipo == 'C':
    if kwh <= 1000:
        preco_kwh = 0.55
    else:
        preco_kwh = 0.60

elif tipo == 'I':
    if kwh <= 5000:
        preco_kwh = 0.55
    else:
        preco_kwh = 0.60

else:
    print("Tipo de instalação inválido!")
    valido = False


if valido:
    total = kwh * preco_kwh
    print(f"Total a pagar: R$ {total:.2f}")
```
